```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
gender=[]
for i in dataset['sex']:
  if i=='male':
    gender.append(1)
  else:
    gender.append(2)
gender
```

```python
dataset['sex']=gender
```

```python
dataset
```

```python
bin=[]
for i in dataset['smoker']:
  if i=='yes':
    bin.append(1)
  else:
    bin.append(0)
bin
```

```python
dataset
```

```python
dataset['smoker']=bin
```

```python
dataset
```

```python
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
A_train,A_test,B_train,B_test,C_train,C_test,D_train,D_test,E_train,E_test,y_train,y_test=train_test_split(dataset['age'],dataset['sex'],dataset['bmi'],dataset['children'],dataset['smoker'],dataset['charges'],test_size=0.20,random_state=0)
```

```python
import numpy as np
```

```python
A_train=A_train.values.reshape(-1,1)
B_train=B_train.values.reshape(-1,1)
C_train=C_train.values.reshape(-1,1)
D_train=D_train.values.reshape(-1,1)
E_train=E_train.values.reshape(-1,1)
y_train=y_train.values.reshape(-1,1)
```

```python
x_train=np.concatenate((A_train,B_train,C_train,D_train,E_train),axis=1)
```

```python
model=regressor.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
A_test=A_test.values.reshape(-1,1)
B_test=B_test.values.reshape(-1,1)
C_test=C_test.values.reshape(-1,1)
D_test=D_test.values.reshape(-1,1)
E_test=E_test.values.reshape(-1,1)
```

```python
x_test=np.concatenate((A_test,B_test,C_test,D_test,E_test),axis=1)
```

```python
y_pred=model.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
mse
```

```python
weight=model.coef_[0]
weight
```

```python
bias=model.intercept_
bias
```
